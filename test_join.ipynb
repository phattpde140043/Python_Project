```python
import pandas as pd
from IPython.display import display, HTML
import requests
```

```python
#GDP to√†n c·∫ßu nƒÉm 2014, k√®m m√£ qu·ªëc gia (ISO code).
df = pd.read_csv("https://raw.githubusercontent.com/plotly/datasets/master/2014_world_gdp_with_codes.csv")

# G·ªçi API
url = "https://restcountries.com/v3.1/all?fields=name,latlng,cca3"
headers = {"User-Agent": "Mozilla/5.0"}
response = requests.get(url, headers=headers)
# Parse JSON
data = response.json()
# Chuy·ªÉn sang DataFrame
records = []
for country in data:
    if "name" in country and "latlng" in country:
        name = country["name"]["common"]
        lat, lon = country["latlng"]
        code= country["cca3"]
        records.append({
            "country": name,
            "latitude": lat,
            "longitude": lon,
            "code": code
        })
df_5 = pd.DataFrame(records)

# Chu·∫©n h√≥a t√™n c·ªôt v·ªÅ c√πng ki·ªÉu
df.rename(columns={"CODE": "code", "COUNTRY": "country_name", "GDP (BILLIONS)": "gdp_billion"}, inplace=True)
# Merge theo m√£ qu·ªëc gia ISO3
merged_df = pd.merge(df, df_5, on="code", how="left")
display(HTML(f"""
<div style="overflow-x: auto; overflow-y: auto; height:400px; width:100%;">
{merged_df.to_html()}
</div>
"""))
```
